# **Scatter Plot**

```python
# plotting histogram of each data to find pattern
def plot_histograms(df):
  fig, axes = plt.subplots(nrows=4, ncols=5, figsize=(30, 20))
  # Iterate over columns and plot histograms
  for i, col in enumerate(df.columns):
    row = i // 5
    col_index = i % 5
    axes[row, col_index].hist(df[col])
    axes[row, col_index].set_title(col)
  plt.tight_layout()
  plt.show()

plot_histograms(df)
```
